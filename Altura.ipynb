```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
pben = pd.read_excel("Arquivos de dados/BD das turmas.xlsx")
```

```python
faixa = np.linspace(1.54,2,15)
def formatação():
    plt.xlabel("Altura dos Alunos")
    plt.ylabel("Quantidade de Alunos")
```

```python
Turma1 = pben.loc[(pben["Turma"] == 1), ["Altura"]].sort_values("Altura", ascending=False)
h1 = Turma1["Altura"].tolist()
Turma2 = pben.loc[(pben["Turma"] == 2), ["Altura"]].sort_values("Altura", ascending=False)
h2 = Turma2["Altura"].tolist()
Turma3 = pben.loc[(pben["Turma"] == 3), ["Altura"]].sort_values("Altura", ascending=False)
h3 = Turma3["Altura"].tolist()
Turma4 = pben.loc[(pben["Turma"] == 4), ["Altura"]].sort_values("Altura", ascending=False)
h4 = Turma4["Altura"].tolist()
CA = pben.loc[0: , ["Altura"]].sort_values("Altura", ascending=False)
hCA = CA["Altura"].tolist()
```

```python
alturas = [h1, h2, h3, h4]
cores = ["blue", "red", "green", "blue"]
```

```python
for i in range(4):
    plt.figure(f"{i+1} Turma")
    plt.hist(alturas[i] ,bins = faixa , color=cores[i], edgecolor='black')
    formatação()
    plt.title(f"Distribuição por altura dos Alunos da Turma {i+1}")

plt.figure("Corpo de Aspirantes")
plt.hist(hCA ,bins = faixa , color="gold", edgecolor='black')
formatação()
plt.title("Distribuição por altura dos Alunos do CA")
```
